```python
from langchain_chroma import Chroma
from langchain_anthropic import ChatAnthropic
from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEmbeddings

# Читаем файл
with open("/Users/vladimirrusakov/Desktop/catalog.txt", "r", encoding="utf-8") as f:
    lines = [line.strip() for line in f if line.strip()]

# Превращаем каждую строку в Document
docs = [Document(page_content=line) for line in lines]

# Создаём векторную базу
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
db = Chroma.from_documents(docs, embeddings)

print(f"Загружено документ
```

```python
from langchain_anthropic import ChatAnthropic
from langchain_core.messages import HumanMessage
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv("ANTHROPIC_API_KEY")
client = ChatAnthropic(
    model="claude-opus-4-5",
    api_key=api_key,
    max_tokens=1024
)

while True:
    query = input("Какие есть модели?")
    if query.lower() in ["выход", "exit", "quit"]:
        print("Выход из программы.")
        break

    # Ищем релевантные документы
    found_docs = db.similarity_search(query, k=3)
    context = "\n".join([doc.page_content for doc in found_docs])

    # Передаём найденное в LLM
    response = client.invoke([
        HumanMessage(content=f"Вот каталог моделей:\n{context}\n\nВопрос: {query}")
    ])
    print("Ответ от модели:", response.content)
```
