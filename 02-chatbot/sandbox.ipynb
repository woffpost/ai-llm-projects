```python
import anthropic
import asyncio
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv("ANTHROPIC_API_KEY")

client = anthropic.AsyncAnthropic(api_key=api_key)
# model_name = input("Введи название модели: ")



async def generate_caption(model):
    try:
        message = await client.messages.create(
            model="claude-opus-4-5",
            max_tokens=1024,
            system="Ты эксперт по советским масштабным моделям 1:43. Пишешь короткие живые подписи для Instagram на русском языке с эмодзи и хэштегами.",
            messages=[{"role": "user", "content": f"Напиши подпись для модели: {model}"}]
        )
        return f"=== {model} ===\n{message.content[0].text}\n"
    except Exception as e:
        print(f"Ошибка при генерации подписи для модели {model}: {e}")

async def main():
    with open("/Users/vladimirrusakov/Desktop/models.txt", "r", encoding="utf-8") as f:
        models = [line.strip() for line in f if line.strip()]
    results = await asyncio.gather(*[generate_caption(model) for model in models])
    results = [r for r in results if r]

    with open("/Users/vladimirrusakov/Desktop/results.txt", "w", encoding="utf-8") as f:
        f.write("\n".join(results))

    print('Готово! Обработано моделей', len(results))

await main()
```
